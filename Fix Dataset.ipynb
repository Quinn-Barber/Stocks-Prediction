```python
import math
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
close_dataset = pd.read_csv("Predictions/Close Price Prediction.csv")
open_dataset = pd.read_csv("Predictions/Open Price Prediction.csv")
low_dataset = pd.read_csv("Predictions/Low Price Prediction.csv")
high_dataset = pd.read_csv("Predictions/High Price Prediction.csv")
```

```python
length = len(close_dataset)
```

```python
for index in range(length):
    c, o, l, h = close_dataset.at[index, "Close"], open_dataset.at[index, "Open"], low_dataset.at[index, "Low"], high_dataset.at[index, "High"]
    l = min(c, o, l, h)
    h = max(c, o, l, h)
    
    low_dataset.loc[index, "Low"] = l
    high_dataset.loc[index, "High"] = h
    
```

```python
df = pd.concat([open_dataset, high_dataset, low_dataset, close_dataset], axis=1)
df.tail(20)
```

```python
df.to_csv("Predictions\Overall_Predictions.csv") 
```
